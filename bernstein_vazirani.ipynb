```python
from qalgo import bernstein_vazirani
from qiskit import *
from qiskit.visualization import plot_histogram
```

```python
oracle = bernstein_vazirani.BernsteinVaziraniOracle(5, 0b01011)
qc = bernstein_vazirani.BernsteinVaziraniCircuit(5, oracle)
qc.draw('mpl')
```

```python
simulator = Aer.get_backend('qasm_simulator')
job = execute(qc, simulator)
plot_histogram(job.result().get_counts())
```
