```python
from qalgo import deutsh_jozsa, deutsh
from qiskit import *
from qiskit.visualization import plot_histogram
```

```python
qc0 = deutsh.DeutshCircuit(deutsh.DeutshConstantOracle(0))
qc0.draw('mpl')
```

```python
qc = deutsh_jozsa.DeutshJozsaCircuit(2,
    deutsh_jozsa.DeutshJozsaBalancedOracle(2, 0b01))
qc.draw('mpl')
```

```python
balanced_oracle_circuit = deutsh_jozsa.DeutshJozsaBalancedOracleCircuit(4, 0b0101)
balanced_oracle_circuit.draw('mpl')
```

```python
simulator = Aer.get_backend('qasm_simulator')
job = execute(qc, simulator)
plot_histogram(job.result().get_counts())
```

```python
qc2 = deutsh_jozsa.DeutshJozsaCircuit(2,
    deutsh_jozsa.DeutshJozsaConstantOracle(2, True))
qc2.draw('mpl')
```

```python
simulator = Aer.get_backend('qasm_simulator')
job = execute(qc2, simulator)
plot_histogram(job.result().get_counts())
```
